```python
def prod_string(str):
    prod = 1 
    for i in range(13):
        prod *= int(str[i])

    if len(str) == 13:
        return prod
    
    i = 0
    j = 12
    max_prod  = prod
    while j+1 < len(str):
        prod //= int(str[i])
        prod *= int(str[j+1])
        i += 1
        j += 1
        if prod > max_prod:
            max_prod = prod
    
    return max_prod
    
```
